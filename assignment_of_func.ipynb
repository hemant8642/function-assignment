**1. What is the difference between a function and a method in Python?**

**ANS**
In Python, functions and methods are both blocks of code that can be executed multiple times from different parts of a program. However, there are key differences between them:

**Functions**

- A function is a self-contained block of code that takes arguments, performs some operations, and returns values.
- Functions are defined using the def keyword and can be called by their name.
- Functions are not part of a class and do not have access to the class's attributes.

**Methods**

- A method is a function that belongs to a class and is used to perform operations on the class's attributes.
- Methods are also defined using the def keyword but are inside a class definition.
- Methods have access to the class's attributes and can modify them.


```python
#Example


# Function
def greet(name):
    print(f"Hello, {name}!")

# Calling the function
greet("John")

# Class with a method
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print(f"Hello, my name is {self.name}!")

# Creating an instance of the class
john = Person("John")

# Calling the method
john.greet()
```

**6. Explain the concept of generators in Python and how they are defined**

**ANS**

In Python, a generator is a special type of function that can be used to generate a sequence of values instead of computing them all at once and returning them in a list, for example.

Generators vs. Functions

The key difference between a generator and a function is that a generator uses the yield keyword to produce a value, whereas a function uses the return keyword to produce a value.

Why Generators?

Generators are useful when you need to generate a large sequence of values, but you don't need to store all of them in memory at once. This can be particularly useful for large datasets or for infinite sequences.

Defining a Generator

A generator is defined using a function definition, but instead of using the return keyword, you use the yield keyword to produce a value.

Here's an example of a simple generator:


```python
def infinite_sequence():
    num = 0
    while True:
        yield num
        num += 1

# Create a generator object
gen = infinite_sequence()

# Print the first 10 values
for _ in range(10):
    print(next(gen))
```

**7. What are the advantages of using generators over regular functions?**

**ANS**

Generators have several advantages over regular functions:

Advantages of Generators

1. Memory Efficiency: Generators use significantly less memory than regular functions because they don't store all the values in memory at once. Instead, they generate values on the fly.
2. Lazy Evaluation: Generators only evaluate the next value when asked, which can be useful when working with large datasets or infinite sequences.
3. Flexibility: Generators can be used to implement cooperative multitasking, which allows for efficient and lightweight concurrency.
4. Improved Performance: Generators can be faster than regular functions because they avoid the overhead of creating and returning lists.


```python
#Example: Fibonacci Sequence

#Here's an example that demonstrates the advantages of generators over regular functions:
# Generator function
def fibonacci_generator():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b
fib_gen = fibonacci_generator()
for _ in range(10):
    print(next(fib_gen))
```

**9. Explain the purpose and usage of the `map()` function in Python.**

**ANS**

The map() function in Python is a built-in function that applies a given function to each item of an iterable (such as a list, tuple, or string) and returns a map object.

**Purpose**

The purpose of the map() function is to:

1. Apply a transformation to each element of an iterable.
2. Create a new iterable with the transformed elements.
3. Avoid explicit loops and make the code more concise.

**Usage**

The map() function takes two arguments:

1. function: The function to be applied to each element of the iterable.
2. iterable: The iterable to be transformed.

The syntax is as follows:

map(function, iterable)

```python
#Example

#Here's an example that demonstrates the usage of the map() function:

# Define a function to square a number
def square(x):
    return x ** 2
# Create a list of numbers
numbers = [1, 2, 3, 4, 5]

# Use map() to square each number in the list
squared_numbers = list(map(square, numbers))

# Print the result
print(squared_numbers)
```

**10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

**ANS**

**1. map() function**

The map() function applies a given function to each item of an iterable (such as a list, tuple, or string) and returns a map object.

- Syntax: map(function, iterable)


```python
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(squared_numbers)
```

**2. reduce() function**

The reduce() function applies a given function to the items of an iterable, going from left to right, so as to reduce the iterable to a single output.

- Syntax: reduce(function, iterable)

```python
#- Example:

import functools
numbers = [1, 2, 3, 4, 5]
sum_of_numbers = functools.reduce(lambda x, y: x + y, numbers)
print(sum_of_numbers)
```

**filter() function**

The filter() function constructs an iterator from elements of an iterable for which a function returns True.

- Syntax: filter(function, iterable)

```python
# Example:

numbers = [1, 2, 3, 4, 5]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)
```

the list.**

```python
def sum_even_numbers(numbers):

    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum

# Example usage:
numbers = [1, 2, 3, 4, 5, 6]
result = sum_even_numbers(numbers)
print("The sum of even numbers is:", result)
```

**4. Write a Python function that checks if a given number is prime or not from 1 to 200.**

```python
def check_prime(n):

    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

def print_primes_up_to_n(n):

    for i in range(1, n + 1):
        if check_prime(i):
            print(i)

# Example usage:
print_primes_up_to_n(200)
```

**5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.**

```python
#Here is a Python iterator class that generates the Fibonacci sequence up to a specified number of terms:


class FibonacciIterator:

    def __init__(self, n):

        self.n = n
        self.a, self.b = 0, 1
        self.current_term = 0

    def __iter__(self):

        return self

    def __next__(self):
      if self.current_term >= self.n:
            raise StopIteration
      result = self.a
      self.a, self.b = self.b, self.a + self.b
      self.current_term += 1
      return result
# Example usage:
fib_iterator = FibonacciIterator(10)
for term in fib_iterator:
    print(term)

```

**6. Write a generator function in Python that yields the powers of 2 up to a given exponent.**

```python
def powers_of_two(exponent):

    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
exponent = 10
print("Powers of 2 up to exponent", exponent, ":")
for power in powers_of_two(exponent):
    print(power)
```

**7. Implement a generator function that reads a file line by line and yields each line as a string.**

```python
def read_file_line_by_line(filename):

    try:
        with open(filename, 'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"File '{filename}' not found.")

# Example usage:
filename = 'example.txt'
print(f"Reading file '{filename}':")
for line in read_file_line_by_line(filename):
    print(line)
```

**9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit**

```python
# Define a function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Define a list of temperatures in Celsius
celsius_temperatures = [0, 10, 20, 30, 40]

# Use map() to convert the list of temperatures from Celsius to Fahrenheit
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Print the converted temperatures
print("Celsius Temperatures:", celsius_temperatures)
print("Fahrenheit Temperatures:", fahrenheit_temperatures)
```

**10. Create a Python program that uses `filter()` to remove all the vowels from a given string.**

```python
# Define a function to check if a character is a vowel
def is_vowel(char):
    vowels = 'aeiouAEIOU'
    return char in vowels

# Define a string
input_string = "Hello, World!"

# Use filter() to remove all the vowels from the string
vowel_free_string = "".join(filter(lambda char: not is_vowel(char), input_string))

# Print the result
print("Original String:", input_string)
print("Vowel-Free String:", vowel_free_string)
```
